```python
df = pd.read_csv(file_path)
```

- Report the number of rows and columns


```python
num_rows = df.shape[0]
num_columns = df.shape[1]

print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

- Report the min, max, and mean of Height, Weight and BMI columns


```python
#1 Height discriptive statistics
min_height = df['Height_(cm)'].min()
max_height = df['Height_(cm)'].max()
mean_height = df['Height_(cm)'].mean()

print("Height")
print(f"Minimum height: {min_height}")
print(f"Maximum height: {max_height}")
print(f"Mean height: {mean_height}")

#2 Weight discriptive statistics
min_weight = df['Weight_(kg)'].min()
max_weight = df['Weight_(kg)'].max()
mean_weight = df['Weight_(kg)'].mean()

print('\nWeight')
print(f"Minimum weight: {min_weight}")
print(f"Maximum weight: {max_weight}")
print(f"Mean weight: {mean_weight}")

#3 BMI discriptive statistics
min_BMI = df['BMI'].min()
max_BMI = df['BMI'].max()
mean_BMI = df['BMI'].mean()

print('\nBMI')
print(f"Minimum BMI: {min_BMI}")
print(f"Maximum BMI: {max_BMI}")
print(f"Mean BMI: {mean_BMI}")
```

- Report the number of column names, null values, and their data types


```python
# 1. Number of column names
num_columns = len(df.columns)
print(f"Number of column names: {num_columns}")

# 2. Null values (NaN) in each column
null_values_per_column = df.isnull().sum()
print("\nNull values per column:")
print(null_values_per_column)

# 3. Data types of each column
column_data_types = df.dtypes
print("\nData types per column:")
print(column_data_types)
```
